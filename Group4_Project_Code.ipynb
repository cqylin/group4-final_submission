```python
#Check plausible values
#Continuous vars: Identify outliers using boxplots, remove outliers outside of IQR
#[Age_at_Exam, sBP, BMI, LDL, HDL, A1c, TG, FBS]
continuous_vars = ["Age_at_Exam", "BMI", "LDL", "HDL", "A1c", "TG", "FBS"]

# Function to plot boxplots
def plot_boxplots(df, title):
    plt.figure(figsize=(12, 8))
    for i, col in enumerate(continuous_vars, 1):
        if col in df.columns:
            plt.subplot(2, 4, i)
            sns.boxplot(y=df[col])
            plt.title(col)
    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

# Function to remove outliers using IQR
def remove_outliers(df, cols):
    for col in cols:
        if col in df.columns:  # Ensure the column exists
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1

            # Define lower and upper bounds for outliers
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Filter data to keep only values within the bounds
            df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    return df

# Plot boxplots before removing outliers
plot_boxplots(df, "Boxplots Before Outlier Removal")

# Apply the function to remove outliers
df = remove_outliers(df, continuous_vars)

# Plot boxplots after removing outliers
plot_boxplots(df, "Boxplots After Outlier Removal")
```

## Logistic Regression


```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import sem, t
from sklearn.utils import resample
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix
```

```python
# Logistic Regression
# List of variables that are going to be included
X_lr = df2[['Age_at_Exam', 'BMI', 'LDL', 'HDL', 'TG', 'Sex', 'Depression', 'OA', 'COPD', 'Anxiety',
            '2ndline-antidiabetic', 'anticholesterol', 'antihtn', 'corticosteroid', '1stlineantidiabetic']]

y_lr = df2['HTN']

# Step 1
# Split the data into training and testing sets
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_lr, y_lr, test_size=0.2, random_state=42, stratify=y_lr
)

# Step 2
# The pipeline ensures that scaling is applied within each fold of cross-validation
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),  # Standardize the data
    ('model', LogisticRegression(max_iter=1000))  # Logistic Regression model
])

# Step 3
# Hyperparameter optimization
log_reg_lr = LogisticRegression()
param_grid_lr = {
    'model__C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'model__penalty': ['l1', 'l2'],  # Type of regularization
    'model__solver': ['liblinear', "lbfgs", "saga"]  # Solver that works with both L1 and L2 penalties
}

grid_search_lr = GridSearchCV(
    estimator=pipeline_lr,
    param_grid=param_grid_lr,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),  # 5-fold CV
    scoring='f1',  # Metric to optimize
    n_jobs=-1,  # Use all available CPU cores
    verbose=2  # Show progress
)

# Step 4
# Fit GridSearchCV on the training data
grid_search_lr.fit(X_train_lr, y_train_lr)
# Get the best hyperparameters
best_params_lr = grid_search_lr.best_params_
print(f"Best Hyperparameters: {best_params_lr}")

# Step 5
# Evaluate the best model on the test set
best_model_lr = grid_search_lr.best_estimator_  # Get the best model
y_test_pred_lr = best_model_lr.predict(X_test_lr)  # Predict on the test set
y_test_proba_lr = best_model_lr.predict_proba(X_test_lr)[:, 1]  # Probabilities for the positive class
```

```python
# Step 8: Reporting the final results
def bootstrap_ci_lr(metric_fn, y_true_lr, y_pred_lr, y_proba_lr=None, n_bootstrap_lr=1000, confidence_lr=0.95):
    metric_scores_lr = []
    for _ in range(n_bootstrap_lr):
        # Resample with replacement
        indices_lr = resample(range(len(y_true_lr)), replace=True, n_samples=len(y_true_lr))
        y_true_sample_lr = np.array(y_true_lr)[indices_lr]
        y_pred_sample_lr = np.array(y_pred_lr)[indices_lr]

        # Compute metric
        if y_proba_lr is not None:
            y_proba_sample_lr = np.array(y_proba_lr)[indices_lr]
            score_lr = metric_fn(y_true_sample_lr, y_proba_sample_lr)  # Used for AUC
        else:
            score_lr = metric_fn(y_true_sample_lr, y_pred_sample_lr)  # Used for precision, recall, F1

        metric_scores_lr.append(score_lr)

    # Compute confidence interval
    lower_lr = np.percentile(metric_scores_lr, (1 - confidence_lr) / 2 * 100)
    upper_lr = np.percentile(metric_scores_lr, (1 + confidence_lr) / 2 * 100)

    return np.mean(metric_scores_lr), lower_lr, upper_lr

# Compute classification metrics with 95% CI
auc_mean_lr, auc_lower_lr, auc_upper_lr = bootstrap_ci_lr(roc_auc_score, y_test_lr, y_test_pred_lr, y_test_proba_lr)
precision_mean_lr, precision_lower_lr, precision_upper_lr = bootstrap_ci_lr(precision_score, y_test_lr, y_test_pred_lr)
recall_mean_lr, recall_lower_lr, recall_upper_lr = bootstrap_ci_lr(recall_score, y_test_lr, y_test_pred_lr)
f1_mean_lr, f1_lower_lr, f1_upper_lr = bootstrap_ci_lr(f1_score, y_test_lr, y_test_pred_lr)

# Report Classification Report
print("Classification Report:")
print(classification_report(y_test_lr, y_test_pred_lr))

# Report Confusion Matrix
conf_matrix_lr = confusion_matrix(y_test_lr, y_test_pred_lr)
print("Confusion Matrix:")
print(conf_matrix_lr)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_lr, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Print AUC with CI
print("Test Fold Performance:")
print(f"AUC: {auc_mean_lr:.4f} (95% CI: {auc_lower_lr:.4f} - {auc_upper_lr:.4f})")
print(f"Precision: {precision_mean_lr:.4f} (95% CI: {precision_lower_lr:.4f} - {precision_upper_lr:.4f})")
print(f"Recall: {recall_mean_lr:.4f} (95% CI: {recall_lower_lr:.4f} - {recall_upper_lr:.4f})")
print(f"F1 Score: {f1_mean_lr:.4f} (95% CI: {f1_lower_lr:.4f} - {f1_upper_lr:.4f})")

# Plot ROC curve
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test_lr, y_test_proba_lr)
plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, label=f'Test AUC = {auc_mean_lr:.4f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Step 8: Reporting the final results for Neural Network
def bootstrap_ci_nn(metric_fn, y_true_nn, y_pred_nn, y_proba_nn=None, n_bootstrap_nn=1000, confidence_nn=0.95):
    metric_scores_nn = []
    for _ in range(n_bootstrap_nn):
        # Resample with replacement
        indices_nn = resample(range(len(y_true_nn)), replace=True, n_samples=len(y_true_nn))
        y_true_sample_nn = np.array(y_true_nn)[indices_nn]
        y_pred_sample_nn = np.array(y_pred_nn)[indices_nn]

        # Compute metric
        if y_proba_nn is not None:
            y_proba_sample_nn = np.array(y_proba_nn)[indices_nn]
            score_nn = metric_fn(y_true_sample_nn, y_proba_sample_nn)  # Used for AUC
        else:
            score_nn = metric_fn(y_true_sample_nn, y_pred_sample_nn)  # Used for precision, recall, F1

        metric_scores_nn.append(score_nn)

    # Compute confidence interval
    lower_nn = np.percentile(metric_scores_nn, (1 - confidence_nn) / 2 * 100)
    upper_nn = np.percentile(metric_scores_nn, (1 + confidence_nn) / 2 * 100)

    return np.mean(metric_scores_nn), lower_nn, upper_nn

# Compute classification metrics with 95% CI
auc_mean_nn, auc_lower_nn, auc_upper_nn = bootstrap_ci_nn(roc_auc_score, y_test_nn, y_test_pred_nn, y_test_proba_nn)
precision_mean_nn, precision_lower_nn, precision_upper_nn = bootstrap_ci_nn(precision_score, y_test_nn, y_test_pred_nn)
recall_mean_nn, recall_lower_nn, recall_upper_nn = bootstrap_ci_nn(recall_score, y_test_nn, y_test_pred_nn)
f1_mean_nn, f1_lower_nn, f1_upper_nn = bootstrap_ci_nn(f1_score, y_test_nn, y_test_pred_nn)

# Report Classification Report
print("Classification Report:")
print(classification_report(y_test_nn, y_test_pred_nn))

# Report Confusion Matrix
conf_matrix_nn = confusion_matrix(y_test_nn, y_test_pred_nn)
print("Confusion Matrix:")
print(conf_matrix_nn)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Print AUC with CI
print("Test Fold Performance:")
print(f"AUC: {auc_mean_nn:.4f} (95% CI: {auc_lower_nn:.4f} - {auc_upper_nn:.4f})")
print(f"Precision: {precision_mean_nn:.4f} (95% CI: {precision_lower_nn:.4f} - {precision_upper_nn:.4f})")
print(f"Recall: {recall_mean_nn:.4f} (95% CI: {recall_lower_nn:.4f} - {recall_upper_nn:.4f})")
print(f"F1 Score: {f1_mean_nn:.4f} (95% CI: {f1_lower_nn:.4f} - {f1_upper_nn:.4f})")

# Plot ROC curve
fpr_nn, tpr_nn, thresholds_nn = roc_curve(y_test_nn, y_test_proba_nn)
plt.figure(figsize=(8, 6))
plt.plot(fpr_nn, tpr_nn, label=f'Test AUC = {auc_mean_nn:.4f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
